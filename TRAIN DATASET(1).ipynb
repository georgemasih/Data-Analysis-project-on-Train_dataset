### Changing the data_type of Date_of_Journey(column) into datetime data_type

```python
s["Date_of_Journey"]=pd.to_datetime(s["Date_of_Journey"])
```

### Create new column of month,day,year 

```python
s["month"]=s["Date_of_Journey"].apply(lambda x:x.month)
s["date"]=s["Date_of_Journey"].apply(lambda x:x.day)
s["year"]=s["Date_of_Journey"].apply(lambda x:x.year)
```

```python
s
```

### Cleaning Arrival_time column

```python
s["Arrival_Time"]=s["Arrival_Time"].str.replace("Mar"," ")
s["Arrival_Time"]=s["Arrival_Time"].str.replace("Apr"," ")
s["Arrival_Time"]=s["Arrival_Time"].str.replace("May"," ")
s["Arrival_Time"]=s["Arrival_Time"].str.replace("Jun"," ")
```
